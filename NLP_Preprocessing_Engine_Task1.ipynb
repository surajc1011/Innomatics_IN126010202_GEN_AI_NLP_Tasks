Task 1: Conceptual Understanding
1. What is the difference between "Love" and "love" in NLP?

In NLP, "Love" and "love" are treated as different tokens if case normalization is not applied. "Love" may appear at the beginning of a sentence or as a proper noun, while "love" is typically a common noun or verb. Without converting text to lowercase, the model may treat them as separate words, increasing vocabulary size and reducing efficiency. Therefore, most preprocessing pipelines convert text to lowercase to ensure consistency.

2. What happens if stopwords are not removed?

If stopwords (such as "is", "the", "and") are not removed, they can:

Add noise to the data since they occur very frequently but carry little meaning
Increase computational cost due to a larger number of tokens
Reduce model performance by diluting important keywords

However, in some cases, keeping stopwords is important for understanding context.

3. Provide two real-world scenarios where removing stopwords can be harmful
Sentiment Analysis
Example:
"I do not like this product"
Removing "not" changes the meaning completely and leads to incorrect sentiment prediction.
Question Answering / Search Queries
Example:
"What is the best phone?"
Removing stopwords like "what" and "is" may distort the intent of the query, making it harder for systems to understand the question.
4. Explain the difference between stemming and lemmatization
Stemming is a rule-based process that removes suffixes from words to get a base form. It is faster but less accurate.
Example:
"running" → "run"
"studies" → "studi" (may not be a valid word)
Lemmatization uses vocabulary and grammar rules to convert words into their correct base or dictionary form. It is slower but more accurate.
Example:
"running" → "run"
"better" → "good"

Key Difference:
Stemming may produce non-meaningful words, while lemmatization always returns meaningful root words.

```python
#Task 2:Build Advanced Preprocessing Function
import re

def preprocess_text(text):

    # 1. Convert to lowercase
    text = text.lower()

    # 2. Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # 3. Remove email patterns
    text = re.sub(r'\S+@\S+', '', text)

    # 4. Remove numbers
    text = re.sub(r'\d+', '', text)

    # 5. Handle repeated characters (soooo → soo)
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # 6. Remove special characters / punctuation
    text = re.sub(r'[^\w\s]', '', text)

    # 7. Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # 8. Tokenization (split into words)
    tokens = text.split()

    # 9. Remove very short tokens (length ≤ 2), but keep "no" and "not"
    filtered_tokens = []
    for word in tokens:
        if len(word) > 2 or word in ["no", "not"]:
            filtered_tokens.append(word)

    return filtered_tokens
sample_text = "WOW!!! This is soooo goooood!!! Visit http://example.com now or mail me at test@gmail.com. I have 2 dogs 😄"

result = preprocess_text(sample_text)

print("Original Text:\n", sample_text)
print("\nProcessed Tokens:\n", result)
```

```python
#Task 4: Token Analytics
# Function to compute analytics
def token_analytics(tokens):
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))

    if total_tokens > 0:
        avg_length = sum(len(word) for word in tokens) / total_tokens
    else:
        avg_length = 0

    return total_tokens, unique_tokens, round(avg_length, 2)


# Run analytics on test sentences
results = []

for i, sentence in enumerate(test_sentences, 1):
    tokens = preprocess_text(sentence)
    cleaned_sentence = " ".join(tokens)

    total, unique, avg_len = token_analytics(tokens)

    results.append({
        "id": i,
        "original": sentence,
        "tokens": tokens,
        "cleaned": cleaned_sentence,
        "total": total,
        "unique": unique,
        "avg_len": avg_len
    })

    print(f"\n--- Test Case {i} ---")
    print("Cleaned Tokens:", tokens)
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Avg Token Length:", avg_len)
```

```python
#Task 5: Frequency Analysis
from collections import Counter

all_tokens = []

for sentence in test_sentences:
    tokens = preprocess_text(sentence)
    all_tokens.extend(tokens)

print("All Tokens:\n", all_tokens)
# Count frequency
freq = Counter(all_tokens)

# Top 10 most frequent words
top_10 = freq.most_common(10)

# Top 5 least frequent words
least_5 = freq.most_common()[:-6:-1]

print("\nTop 10 Most Frequent Words:")
for word, count in top_10:
    print(word, ":", count)

print("\nTop 5 Least Frequent Words:")
for word, count in least_5:
    print(word, ":", count)
```

```python
#Task 6: Build Full Pipeline
import re

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()

    tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]

    return tokens

test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

def full_pipeline(text_list):

    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens = preprocess_text(text)

        all_tokens.extend(tokens)
        clean_sentences.append(" ".join(tokens))

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

output = full_pipeline(test_sentences)

print("TOKENS:\n", output["tokens"])
print("\nCLEAN SENTENCES:\n", output["clean_sentences"])
```
